# Merge, Join, and Concatenate

```python
import pandas as pd
```

```python
df1 = pd.read_csv(r"C:\Users\user\Dropbox\PC\Downloads\LOTR.csv")
df1
```

```python
df2 = pd.read_csv(r"C:\Users\user\Dropbox\PC\Downloads\LOTR 2.csv")
df2
```

## merge

```python
# merge, default -> inner join
# df1 as left dataframe, df2 as right dataframe

df1.merge(df2)
```

```python
# merge (inner join), on 'FellowshipID' 

df1.merge(df2, how = 'inner', on = 'FellowshipID')
```

```python
df1.merge(df2, how = 'inner', on = ['FellowshipID', 'FirstName'])
```

```python
# merge, outer join 
# couldn't match -> NaN

df1.merge(df2, how = 'outer')
```

```python
# take everything from dataframe 1, if there is any overlap from dataframe 2, it would also show

df1.merge(df2, how = 'left')
```

```python
# take everything from dataframe 2, if there is any overlap from dataframe 1, it would also show

df1.merge(df2, how = 'right')
```

```python
# take each values from the left dataframe, and compare with the values from the right dataframe

df1.merge(df2, how = 'cross')
```

## join

```python
# join (very similar to merge function)
# merge has more default, works better with column names
# join works better with indexes

df1.join(df2, on = 'FellowshipID', how = 'outer', lsuffix = '_left', rsuffix = '_right')
```

```python
# creat indexes and join on the index

df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix = '_left', rsuffix = '_right')
df4
```

```python
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix = '_left', rsuffix = '_right', how = 'outer')
df4
```

## concat

```python
# concatenate, put one dataframe on top of another

pd.concat([df1,df2])
```

```python
pd.concat([df1,df2], join = 'inner')
```

```python
# default

pd.concat([df1,df2], join = 'outer')
```

```python
# join on the index

pd.concat([df1,df2], join = 'outer', axis = 1)
```

# Data Cleaning

```python
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\user\Dropbox\PC\Downloads\Customer Call List.xlsx")
df
```

```python
# delete the duplicates

df = df.drop_duplicates()
df
```

```python
df = df.drop(columns = 'Not_Useful_Column')
df
```

```python
# clean last name column
# strip -> take out the values on the both end side

df['Last_Name'] = df['Last_Name'].str.lstrip('...')
df
```

```python
df['Last_Name'] = df['Last_Name'].str.lstrip('/')
df['Last_Name'] = df['Last_Name'].str.rstrip('_')
df
```

```python
# remove all unwanted values at once

df['Last_Name'].str.strip('123._/')
```

```python
# save it to the dataframe

df['Last_Name'] = df['Last_Name'].str.strip('123._/')
```

```python
df
```

```python
# phone numbers with different format -> make all look the same
# make all of them just numeric (only numbers), then re-format it
# replace everything that is not 'a to z, A to Z, 0 to 9' to empty space

df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','', regex = True)
df
```

```python
# make the phone number column to string

df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x)) 
```

```python
# add '-' into phone numbers 

df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
```

```python
df
```

```python
# remove the additional '--' at the end of None value

df["Phone_Number"] = df["Phone_Number"].str.replace('nan--','')
df["Phone_Number"] = df["Phone_Number"].str.replace('Na--','')
df
```

```python
# separate address to three columns

df[['Street_Address', 'State', 'Zip_Code']] = df['Address'].str.split(',', expand = True)
df
```

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes', 'Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No', 'N')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes', 'Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No', 'N')
df
```

```python
df.replace('N/a','')
df.replace('NaN','')
```

```python
# fill up the empty space with blank instead of NaN, N/a, or None
df = df.replace('N/a','')
df = df.fillna('')
df
```

```python
# remove people who chose 'do not contact'

for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y':
        df.drop(x, inplace = True)
df
```

```python
# remove contact without any phone number

for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
        df.drop(x, inplace = True)
df

# another way to drop null values:
# df = df.dropna(subset = 'Phone_Number', inplace = True)
```

```python
# reset the index to reduce the confusion

df = df.reset_index(drop = True)
df
```
